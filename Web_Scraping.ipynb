```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
```

```python
name = []
price = []
instock = []
ratings = []

for j in range(1,10):
  headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)",
    "Accept": "text/html",
    "Accept-Language": "en-US"}

  webpage = requests.get('https://books.toscrape.com/catalogue/page-{}.html'.format(j), headers=headers).text
  soup = BeautifulSoup(webpage, 'lxml')
  books = soup.find_all('li', class_='col-xs-6 col-sm-4 col-md-3 col-lg-3')



  for i in books:
   title = i.find('h3').find('a')['title']
   name.append(title)

   prices = i.find('div', class_ = 'product_price').find('p', class_ = 'price_color').text
   price.append(prices)

   ins = i.find('p', class_ = 'instock availability').text.strip()

   if ins == 'In stock':
      instock.append(True)
   else:
      instock.append(False)

   rating_tag = i.find('p', class_='star-rating')
   rating_classes = rating_tag.get('class')
   rating = rating_classes[1]
   ratings.append(rating)



d= {'Title': name, 'Price':price, 'InStock':instock, 'Ratings':ratings}
df = pd.DataFrame(d)

df
```
